__Задание 1. (1 балл)__

Для каждого датасета обучите решающее дерево с параметрами по умолчанию, предварительно разбив выборку на обучающую и тестовую. Постройте разделящие поверхности (для этого воспользуйтесь функцией `plot_surface`, пример ниже). Посчитайте accuracy на обучающей и тестовой выборках. Сильно ли деревья переобучились?

```python
def plot_surface(clf, X, y, ax=plt):
    plot_step = 0.01
    palette = sns.color_palette(n_colors=len(np.unique(y)))
    cmap = ListedColormap(palette)
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = ax.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)

    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.7, edgecolors=np.array(palette)[y], linewidths=2)
```

```python
def accuracy(pred, y):
    return 1.0 * (np.array(pred) == np.array(y)).sum() / pred.shape[0]

def apply_classic_tree(X, y, i):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42, stratify=y)
    model = DecisionTreeClassifier().fit(X_train, y_train)
    plt.subplot(2, 3, i + 1)
    plt.title("Dataset train " + str(i + 1))
    plot_surface(model, X_train, y_train)
    plt.subplot(2, 3, i + 1 + 3)
    plt.title("Dataset test " + str(i + 1))
    plot_surface(model, X_test, y_test)
    print("Accuracy train =", accuracy(model.predict(X_train), y_train))
    print("Accuracy test  =", accuracy(model.predict(X_test), y_test))
    
plt.figure(figsize=(12, 6))
for i in range(3):
    print("Dataset", i + 1)
    X, y = datasets[i]
    apply_classic_tree(X, y, i)
    print()
```

**Вывод:** Как видно на обучающей выборке точность 1, хотя по картинке видно что разделение не такое точное, что свидетельствует о том что есть какие-то области, которые подстроились под конкретные объекты. На самом деле точность можно оценить по тестовой выборке. Поэтому деревья решений прилично переобучились.

__Задание 2. (1.5 балла)__

Попробуйте перебрать несколько параметров для регуляризации (напр. `max_depth`, `min_samples_leaf`). Для каждого набора гиперпараметров постройте разделяющую поверхность, выведите обучающую и тестовую ошибки. Можно делать кросс-валидацию или просто разбиение на трейн и тест, главное делайте каждый раз одинаковое разбиение, чтобы можно было корректно сравнивать (помните же, что итоговое дерево сильно зависит от небольшого изменения обучающей выборки?). Проследите как меняется разделяющая поверхность и обобщающая способность. Почему так происходит, одинаково ли изменение для разных датасетов?

```python
def apply_tree_with_parameters(X, y, max_depth, min_samples_leaf, ax):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42, stratify=y)
    model = DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf).fit(X_train, y_train)
    train_accuracy = accuracy(model.predict(X_train), y_train)
    test_accuracy = accuracy(model.predict(X_test), y_test)
    ax.set_title(f"Accuracy\ntrain:{train_accuracy:9.4f}, test:{test_accuracy:9.4f}")
    ax.set_xlabel(f"min_samples_leaf:{min_samples_leaf:9.4f}")
    ax.set_ylabel(f"max_depth:{max_depth}")
    plot_surface(model, X, y, ax)

def brute_parameters(X, y):
    max_depths = np.arange(1,14,3)
    min_samples_leafs = np.linspace(0.01, 0.25, num=5)
    fig, axs = plt.subplots(
        max_depths.size, 
        min_samples_leafs.size, 
        figsize=(15, 15), 
        tight_layout=True
    )
    for i in range(max_depths.size):
        for j in range(min_samples_leafs.size):
            apply_tree_with_parameters(X, y, max_depths[i], min_samples_leafs[j], axs[i, j])
```

**Как смотреть на графики:** 
- Сверху **accuracy** разбиения
- Слева значение параметра **max_depth**
- Снизу значение параемтра **min_samples_leaf**

__Ответ:__ 
- Чем более явное разделение классов, тем меньше параметры влияют на обучение модели
- Большая глубина позволяет получить области сложных форм, что зачастую далеко от правды
- Слишком маленькая нижняя граница на количество вершин в листьях также склоняет модель к образованию областей, которые сильно подстраиваются под конкретные объекты
- Видно что глубина дерева сильно влияет на итоговое разбиение. При маленьком min_samples_leaf и большом max_depth происходит явное переобучение
- В целом тенденции похожи на всех трех примерах. Наиболее разумные параметры это max_depth в районе 5 и min_samples_leaf в районе 0.1, то есть 10%

```python
def scatter_feature(feature, target, ax):
    ax.set_xlabel("value", fontsize=15)
    ax.set_yticks([0, 1])
    ax.scatter(feature, target)
    ax.set_ylim(-0.5, 1.5)
    ax.set_xlim(-0.1, 1.1)

def scatter_features(features, target):
    m = features.shape[1]
    fig, axs = plt.subplots(
        1, 
        m, 
        figsize=(20, 3), 
        tight_layout=True
    )
    axs[0].set_ylabel("class", fontsize=15)
    for i in range(m):
        axs[i].set_title(f"Feature {i + 1}", fontsize=15)
        scatter_feature(features[:, i], target, axs[i])
    
scatter_features(data_np[:, :-1], data_np[:, 5])
```

```python
# Скачаем данные
def get_datasets():
    datasets_list = [
        ("mushrooms", "https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data"),
        ("tic_tac_toe", "https://archive.ics.uci.edu/ml/machine-learning-databases/tic-tac-toe/tic-tac-toe.data"),
        ("cars", "https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data"),
        ("nursery", "https://archive.ics.uci.edu/ml/machine-learning-databases/nursery/nursery.data")
    ]
    
    datasets = []
    
    for name, url in datasets_list:
        df = pd.read_csv(url, header=None)
        datasets.append((name, df))
        
    return datasets
```

```python
dfs = get_datasets()
```

```python
# Кодирует целевую переменную на 0 и 1
def prepare_dataset(name, df):
    data = df.to_numpy()
    
    if name == "mushrooms":
        return (data[:, 1:], np.array(data[:, 0] == "e", dtype=int))
    
    if name == "tic_tac_toe":
        return (data[:, :-1], np.array(data[:, -1] == "positive", dtype=int))
    
    if name == "cars":
        return (data[:, :-1], np.array((data[:, -1] == "good") | (data[:, -1] == "vgood"), dtype=int))
    
    if name == "nursery":
        y = data[:, -1]
        return (data[:, :-1], np.array((y == "very_recom") | (y == "priority") | (y == "spec_prior"), dtype=int))
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Применяет LabelEncoder к каждому столбцу
def apply_LabelEncoder(data):
    for i in range(data.shape[1]):
        data[:, i] = LabelEncoder().fit_transform(data[:, i])
    return data
```

```python
prepared_dfs = []

# Обработаем сырые данные в формат numpy, и отделим столбец с целевой переменной с отдельный массив
for name, df in dfs:
    X, y = prepare_dataset(name, df)
    prepared_dfs.append((
        name,
        apply_LabelEncoder(X),
        y
    ))
```

```python
from sklearn.pipeline import Pipeline

def get_models(data):
    return [
        ("simple_real", DecisionTree(np.repeat("real", data.shape[1]))),
        ("simple_categorical", DecisionTree(np.repeat("categorical", data.shape[1]))),
        ("onehot_real", Pipeline(steps=[
            ("scaler", OneHotEncoder(sparse=False)),
            ("model", DecisionTree(np.repeat("real", 1000)))
        ])),
        ("sklearn", DecisionTreeClassifier()),
    ]
```

```python
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.model_selection import cross_val_score

for data_name, X, y in prepared_dfs:
    models = get_models(X)
    for model_name, model in models:
        print(data_name, model_name, cross_val_score(model, X, y, cv=10, scoring=make_scorer(accuracy_score)))
```
