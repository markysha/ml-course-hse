## Задание 1. Реализация градиентного спуска (3.5 балла)

В этом задании вам предстоит написать собственные реализации различных подходов к градиентному спуску с опорой на подготовленные шаблоны в файле  `utils.py`:

**Задание 1.1. (0.5 балла)** Полный градиентный спуск **GradientDescent**:

$$
    w_{k + 1} = w_{k} - \eta_{k} \nabla_{w} Q(w_{k}).
$$

**Задание 1.2. (1 балл)** Стохастический градиентный спуск **StochasticDescent**:

$$
    w_{k + 1} = w_{k} - \eta_{k} \nabla_{w} q_{i_{k}}(w_{k}).
$$ 

$\nabla_{w} q_{i_{k}}(w_{k}) \,$ - это оценка градиента по одному объекту, выбранному случайно.

**Задание 1.3. (1 балл)** Метод Momentum **MomentumDescent**:

$$
    h_0 = 0, \\
    h_{k + 1} = \alpha h_{k} + \eta_k \nabla_{w} Q(w_{k}), \\
    w_{k + 1} = w_{k} - h_{k + 1}.
$$

**Задание 1.4. (1 балл)** Метод Adaptive gradient algorithm **Adagrad**:

$$
    G_0 = 0, \\
    G_{k + 1} = G_{k} + \left(\nabla_{w} Q(w_{k})\right) ^ 2, \\
    w_{k + 1} = w_{k} - \dfrac{\eta_k}{\sqrt{\varepsilon + G_{k}}} \nabla_{w} Q(w_{k}).
$$


Во всех вышеназванных методах мы будем использовать следующую формулу для длины шага:

$$
    \eta_{k} = \lambda \left(\dfrac{s_0}{s_0 + k}\right)^p
$$
На практике достаточно настроить параметр $\lambda$, а остальным выставить параметры по умолчанию: $s_0 = 1, \, p = 0.5.$

Мы будем использовать функцию потерь MSE:

$$
    Q(w) = \dfrac{1}{\ell} \sum\limits_{i=1}^{\ell} (a_w(x_i) - y_i)^2
$$

Все вычисления должны быть векторизованы.

```python
# MomentumDescent

descent = MomentumDescent(lambda_ = lambda_, w0 = w0)

gradient = descent.calc_gradient(X, y)

assert gradient.shape[0] == dimension, 'Gradient failed'

diff = descent.step(X, y, 0)

assert diff.shape[0] == dimension, 'Weights failed'
```

```python
# Adagrad

descent = Adagrad(lambda_ = lambda_, w0 = w0)

gradient = descent.calc_gradient(X, y)

assert gradient.shape[0] == dimension, 'Gradient failed'

diff = descent.step(X, y, 0)

assert diff.shape[0] == dimension, 'Weights failed'
```

## Задание 5. Сравнение методов градиентного спуска (2 балла)

В этом задании вам предстоит сравнить методы градиентного спуска на подготовленных вами данных из предыдущего задания.

* **Задание 5.1. (1.5 балла)** Подберите по валидационной выборке наилучшую длину $\lambda$ шага для каждого метода. Для этого можно сделать перебор по логарифмической сетке, так как нас интересует скорее порядок величины, нежели её точное значение. Сравните качество методов по метрикам MSE и R^2 на обучающей и тестовой выборках, сравните количество итераций до сходимости. Все параметры кроме $\lambda$ стоит выставить равным значениям по умолчанию.

* **Задание 5.2. (0.5 балла)** Постройте график зависимости значения функции ошибки от номера итерации (все методы на одном графике).

Посмотрите на получившиеся результаты. Сравните методы между собой.

```python
GradientDescent.__name__
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score

descent_classes = [
    GradientDescent,
    StochasticDescent,
    MomentumDescent,
    Adagrad,
]

categorical_features = [
    'brand', 
#     'model', 
    'vehicleType', 
    'gearbox', 
    'fuelType',
    'notRepairedDamage',
    'yearOfRegistration',
]

numeric_features = [
    'powerPS',  
    'kilometer', 
]

#DescentName(lambda_ = lambda_, w0 = w0)
alphas = np.logspace(-3, 0.8, 7)
alphas_stats = pd.DataFrame(columns=[
    "alpha",
    "train_mse",
    "validate_mse",
    "test_mse", 
    "train_r2",
    "validate_r2",  
    "test_r2", 
    "descent_name",
    "loss_history",
    "iter",
])


column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), categorical_features),
    ('scaling', StandardScaler(), numeric_features),
#     ('original', NoTransformer(), numeric_features),
], sparse_threshold=0)
# column_transformer.fit(data)

# t_train = column_transformer.transform(train)
# t_validate = column_transformer.transform(validate)

# print(train.head(1))

for alpha in alphas:
    print("Alpha:", alpha)
    
    for descent_class in descent_classes:
        pipeline = Pipeline(steps=[
            ('ohe_and_scaling', column_transformer),
            ('regression', LinearRegression(
                descent = descent_class(lambda_=alpha, w0=0),
                max_iter = 200,
            ))
        ])

        model = pipeline.fit(train, train["log_price"].values)
        train_y = model.predict(train)
        validate_y = model.predict(validate)
        test_y = model.predict(test)
                                
        cur = {
            "alpha": alpha,
            "train_mse": mean_squared_error(train["log_price"], train_y),
            "validate_mse": mean_squared_error(validate["log_price"], validate_y), 
            "test_mse": mean_squared_error(test["log_price"], test_y), 
            "train_r2": r2_score(train["log_price"], train_y),
            "validate_r2": r2_score(validate["log_price"], validate_y),
            "test_r2": r2_score(test["log_price"], test_y), 
            "descent_name": descent_class.__name__,
            "loss_history": pipeline.named_steps["regression"].loss_history,
            "iter": len(pipeline.named_steps["regression"].loss_history)
        }

        alphas_stats = alphas_stats.append(cur, ignore_index=True)
            
```

Найдем лучшую альфу для каждого алгоритма, и посмотрим на ошибку на валидационной выборке

```python
best_alphas_stats = alphas_stats.sort_values("validate_mse").groupby("descent_name").first().reset_index()

best_alphas_stats[["descent_name", "alpha"]]
```

```python
best_alphas_stats[["descent_name", "alpha", "train_mse", "train_r2", "test_mse", "test_r2", "iter"]]
```

Построим график ошибки в зависимости от номера итерации

```python
cur = pd.DataFrame(columns=["descent_name", "iter", "q"])

for index, row in best_alphas_stats.iterrows():
    for i in range(int(row["iter"])):
        add = {
            "descent_name": row["descent_name"],
            "iter": i,
            "q": float(np.log1p(row["loss_history"][i])),
        }
        cur = cur.append(add, ignore_index=True)
    
f, ax = plt.subplots(figsize=(15, 5))
sns.lineplot(
    data=cur, 
    x="iter", y="q",
    hue="descent_name",
    ax = ax
)
ax.set_yscale("log")
ax.set_xlabel('Iteration')
ax.set_ylabel('Error function')
plt.show()
```

**GradientDescent** и **MomentumDescent** получили похожие результаты. Почти одинаковая итоговая ошибка и оптимальная альфа. **MomentumDescent** только несколько быстрее сошелся.

**StohasticDescent** быстро остановился, получив при этом худший результат. Альфа при этом меньше, чем в других алгоритмах.

**Adagrad** получил меньшую ошибку, сделав при этом наибольшее количество итераций. Скорость улучшения ответа однако лучше чем в **GradientDescent** и **MomentumDescent**. Лучшая альфа при этом больше чем в остальных алгоритмах. 

## Задание 6. Сходимость стохастического градиентного спуска в зависимости от размера батча (1 балл)

В этом задании вам предстоит исследовать влияние размера батча на работу стохастического градиентного спуска. 

* Сделайте по несколько запусков (например, k) стохастического градиентного спуска на обучающей выборке для каждого размера батча из списка. Замерьте время и количество итераций до сходимости. Посчитайте среднее и дисперсию этих значений для каждого размера батча.
* Постройте график зависимости количества шагов до сходимости от размера батча.
* Постройте график зависимости времени до сходимости от размера батча.

Посмотрите на получившиеся результаты. Какие выводы можно сделать про подбор размера батча для стохастического градиентного спуска?

```python
import time

batch_sizes = np.arange(5, 200, 15)

StochasticDescent_stats = pd.DataFrame(columns=[
    "batch_size",
    "train_mse",
    "test_mse", 
    "time",
    "k",
    "iter"
])

batch_size = 2
while batch_size < 5000:
    print("batch_size:", batch_size)
    
    descent_class = StochasticDescent
    best_descent_alpha = 0.079433
    
    for k in range(5):
        pipeline = Pipeline(steps=[
            ('ohe_and_scaling', column_transformer),
            ('regression', LinearRegression(
                descent = descent_class(lambda_=best_descent_alpha, w0=0, batch_size=batch_size),
                max_iter = 1000,
            ))
        ])
    
        start = time.time()
        model = pipeline.fit(train, train["log_price"].values)
        end = time.time()
        train_y = model.predict(train)
        test_y = model.predict(test)

        cur = {
            "batch_size": batch_size,
            "k": k,
            "train_mse": mean_squared_error(train["log_price"], train_y), 
            "test_mse": mean_squared_error(test["log_price"], test_y), 
            "iter": len(pipeline.named_steps["regression"].loss_history),
            "time": end - start,
        }

        StochasticDescent_stats = StochasticDescent_stats.append(cur, ignore_index=True)
        
#         print(cur)
    
    batch_size = int(batch_size * 1.5)
```

```python
f, ax = plt.subplots(nrows=3, figsize=(20, 30))
sns.lineplot(
    data=StochasticDescent_stats, 
    x="batch_size", y="time",
    ax = ax[0]
)
ax[0].set_xscale("log")
ax[0].set_xlabel('Batch size')
ax[0].set_ylabel('Time')
sns.lineplot(
    data=StochasticDescent_stats, 
    x="batch_size", y="iter",
    ax = ax[1]
)
ax[1].set_xscale("log")
ax[1].set_xlabel('Batch size')
ax[1].set_ylabel('Iterations')
sns.lineplot(
    data=StochasticDescent_stats, 
    x="batch_size", y="test_mse",
    ax = ax[2]
)
ax[2].set_xscale("log")
ax[2].set_xlabel('Batch size')
ax[2].set_ylabel('Test mse')
plt.show()
```

Кажется, что картина в процессе увеличения размера батча сначала улучшается, достигает лучших результатов, а потом уже разворачивается в обратную сторону.

Так, при очень большом батче, разброса количества итераций почти нет, и точность алгоритма тоже перестает меняться.

С временем работы аналогичная ситуация, лучший период наблюдается с размером батча от 10 до 100.

$$

## Задание 8. Бонус — Реализация метода SAG (2 балла)

В качестве бонуса вам будет следующее задание - напишите собственную реализацию стохастического градиентного спуска по методу SAG в файле `utils.py`. Подробнее прочитать про SAG можно [здесь](https://arxiv.org/pdf/1309.2388.pdf).

Сравните свою реализацию метода SAG с обычным полным градиентным спуском на наших данных. Проведите сравнение аналогично заданию 5. Что вы можете сказать про сходимость этого метода?

```python
%autoreload 2

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
from utils import (
    GradientDescent,
    StochasticAverageGradient,
    StochasticAverageGradient
)

descent_classes = [StochasticAverageGradient, GradientDescent]
categorical_features = ['brand', 'vehicleType', 'gearbox', 'fuelType', 'notRepairedDamage', 'yearOfRegistration']
numeric_features = ['powerPS', 'kilometer']

alphas = np.logspace(-6, 0.8, 10)
sag_stats = pd.DataFrame(columns=[
    "alpha",
    "train_mse", "validate_mse", "test_mse", 
    "train_r2", "validate_r2", "test_r2", 
    "descent_name",
    "loss_history",
    "iter",
])

column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), categorical_features),
    ('scaling', StandardScaler(), numeric_features),
], sparse_threshold=0)

np.random.seed(42)
for alpha in alphas:
    print("Alpha:", alpha)
    
    for descent_class in descent_classes:
        if alpha > 5 and descent_class.__name__ != "StochasticAverageGradient":
            continue
            
        pipeline = Pipeline(steps=[
            ('ohe_and_scaling', column_transformer),
            ('regression', LinearRegression(
                descent = descent_class(
                    lambda_=alpha, 
                    w0=0, 
                    x_shape=train.shape[0]
                ) if descent_class.__name__ == "StochasticAverageGradient" else descent_class(
                    lambda_=alpha, 
                    w0=0, 
                ),
                max_iter=10000,
                tolerance = 0 if descent_class.__name__ == "StochasticAverageGradient" else 1e-2,
#                 max_iter = train.shape[0] if descent_class.__name__ == "StochasticAverageGradient" else 0,
            ))
        ])

        model = pipeline.fit(train, train["log_price"].values)
        train_y = model.predict(train)
        validate_y = model.predict(validate)
        test_y = model.predict(test)
                                
        cur = {
            "alpha": alpha,
            "train_mse": mean_squared_error(train["log_price"], train_y),
            "validate_mse": mean_squared_error(validate["log_price"], validate_y), 
            "test_mse": mean_squared_error(test["log_price"], test_y), 
            "train_r2": r2_score(train["log_price"], train_y),
            "validate_r2": r2_score(validate["log_price"], validate_y),
            "test_r2": r2_score(test["log_price"], test_y), 
            "descent_name": descent_class.__name__,
            "loss_history": pipeline.named_steps["regression"].loss_history,
            "iter": len(pipeline.named_steps["regression"].loss_history)
        }

        sag_stats = sag_stats.append(cur, ignore_index=True)
            
```

Также добавим SAG, в котором за одну итерацию будем обновлять градиент не у одного элемента, а 100

Посмотрим на лучшую альфу

```python
cur = pd.DataFrame(columns=["descent_name", "iter", "q"])

for index, row in best_sag_stats.iterrows():
    for i in range(int(row["iter"])):
        add = {
            "descent_name": row["descent_name"],
            "iter": i,
            "q": float(np.log1p(row["loss_history"][i])),
        }
        cur = cur.append(add, ignore_index=True)
    
f, ax = plt.subplots(figsize=(15, 5))
sns.lineplot(
    data=cur, 
    x="iter", y="q",
    hue="descent_name",
    ax = ax
)
ax.set_yscale("log")
ax.set_xscale("log")
ax.set_xlabel('Iteration')
ax.set_ylabel('Error function')
plt.show()
```

**SAG** делает на порядок больше итераций. Связано это с тем что за одну итерацию градиент меняется не очень сильно, а при большой альфе получаются большие скачки в ненужную сторону, поэтому оптимальная альфа получается маленькая и шаги короткие.

**SAG** потихоньку подстраивается под данные, а обычный **GradientDescent** учитывает сразу все элементы поэтому на начальных итерациях, пока итоговый вектор весов **w** далек от оптимального и градиенты не сильно похожи, получается ухудшение ошибки, а потом с какого-то момента быстрый спуск.
